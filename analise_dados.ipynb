```python
# Esse código calcula o Intervalo Interquartil (IIQ).
#O IIQ é a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1) de um conjunto de dados. É uma medida de dispersão estatística, representando os 50% do meio dos dados.
IQR = Q3 - Q1
IQR
```

```python
# E os limites:
# É usada para calcular o limite superior para identificar outliers. Qualquer valor acima desse limite é considerado um outlier.
lin_superior = Q3 + (1.5 * IQR)
lin_superior
```

```python
# é usada para calcular o limite inferior. Qualquer valor abaixo desse limite é considerado um outlier.
lin_inferior = Q1 - (1.5 * IQR)
lin_inferior
```

```python
from scipy.stats import chi2_contingency
```

```python
# Calcule variáveis discretas e correlação entre as tabelas cruzadas:
def cramer_coeficiente (coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
# Calcule variáveis discretas e correlação entre as tabelas cruzadas:
def cramer_genero(coluna1, coluna2):
    tabela_genero = np.array(pd.crosstab(coluna1, coluna2))
    chi2_2 = chi2_contingency(tabela_genero)[0]
    soma_2 = np.sum(tabela_genero)
    mini_2 = min(tabela_genero.shape) - 1
    cramer_2 = np.sqrt(chi2_2 / (soma_2 * mini_2))
    return cramer_2
```
