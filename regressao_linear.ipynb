```python
# Vamos organizar algumas colunas de cor/raça/etnia para não atrapalhar o resultado do nosso modelo:
lista_retirar = ['Prefiro não informar', 'Outra', 'Indígena']
```

```python
dados = dados[~dados['COR/RACA/ETNIA'].isin(lista_retirar)]
```

```python
dados['NAO_BRANCA'] = dados['COR/RACA/ETNIA'].apply(lambda x: 1 if x != 'Branca' else 0)
```

```python
dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].value_counts()
```

```python
# Na coluna tempo de experiência:
dados['TEMPO_EXPERIENCIA'] = dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].str.extract(r'(\d+)')
```

```python
# Na coluna número de funcionários:
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.replace('.', '')
```

```python
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.extract(r'(\d+)')
```

```python
dados['TEMPO_EXPERIENCIA'] = dados['TEMPO_EXPERIENCIA'].fillna(0)
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts(dropna=False)
```

```python
dados.columns
```

```python
dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].value_counts()
```

```python
dados['INSATISFACAO'] = 0
```

```python
# Vamos criar nossa coluna INSATISFACAO. Após rodar,atribua a coluna 'INSATISFACAO':
dados.loc[dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].notnull(), 'INSATISFACAO'] = dados.loc[dados['Qual o principal motivo da sua insatisfação com a empresa atual?'].notnull(),'Qual o principal motivo da sua insatisfação com a empresa atual?'].apply(lambda x: 1 if 'Salário' in x else 0)
```

```python
dados['INSATISFACAO'].value_counts()
```

```python
dados['NIVEL DE ENSINO'].value_counts()
```

```python
# Na coluna de nível de ensino:
dados['NIVEL DE ENSINO'] = dados['NIVEL DE ENSINO'].apply(lambda x: 0 if x== 'Não tenho graduação formal' else
                                1 if x=='Estudante de Graduação' else
                                2 if x=='Graduação/Bacharelado' else
                                3 if x=='Pós-Graduação' else
                                4 if x=='Mestrado' else
                                5 if x=='Doutorado ou Phd' else -1)
```

```python
# Vamos fazer a seleção das colunas que utilizaremos em nosso modelo e atribuir:
dados = dados[['IDADE', 'GENERO', 'NAO_BRANCA', 'TEMPO_EXPERIENCIA', 'INSATISFACAO', 'SETOR', 'REGIAO ONDE MORA', 'NIVEL DE ENSINO',
      'NUMERO DE FUNCIONARIOS', 'SALARIO', 'NOVO_NIVEL']]
```

```python
dados.columns
```

```python
# Vamos aplicar o get_dummies:
dados = pd.get_dummies(dados, columns=['GENERO', 'SETOR', 'NOVO_NIVEL', 'REGIAO ONDE MORA'], drop_first=True)
```

```python
# Vamos separar nosso conjunto de dados:
x = dados.drop('SALARIO', axis=1)
y = dados['SALARIO']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_scaled =scaler.fit_transform(x_train)
x_test_scaled =scaler.fit_transform(x_test)
```

```python
# Nós vamos utilizar o modelo LinearRegression da biblioteca sklearn e fazemos o import assim:
from sklearn.linear_model import LinearRegression
```

```python
# Criamos o objeto:
model = LinearRegression()
```

```python
# E então enviando os parâmetros de treino na função fit
model.fit(x_train_scaled, y_train)
```

```python
# Vamos fazer as predições do nosso conjunto de teste utilizando a função predict:
# Vamos atribuir o resultado para uma variável que vamos chamar de y_pred:
y_pred = model.predict(x_test_scaled)
```

```python
# Após treinar um modelo de machine learning, vamos avaliar seu desempenho para entender o quão bem ele está fazendo previsões.
# Fazemos o import das funções da biblioteca do sklearn assim:
# Vamos calcular mais uma métrica o MAE. Primeiro, precisamos importar a função. Podemos adicionar à importação anterior e executar em seguida.
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Vamos calcular o erro quadrático médio, o MSE.
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
# Agora sim, fazemos o cálculo do MAE:
mae = mean_absolute_error(y_test, y_pred)
```

```python
mae
```

```python
# Vamos agora calcular o R²
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

```python
# Vamos plotar um gráfico dos valores reais versus os valores preditos pelo modelo.
# Para isso, vamos importar a biblioteca matplotlib:
import matplotlib.pyplot as plt
```

```python
# Na aula, usamos a figura com um tamanho específico (10 polegadas de largura por 6 polegadas de altura). Posteriormente diminuímos para 8,4.
# Plotamos um gráfico de dispersão dos valores reais contra os valores previstos e definimos a transparência dos pontos por 0.5.
# Definimos os rótulos e o título do gráfico.
# Definimos os pontos de início e fim da linha nos eixos X e Y.
# E definimos a cor da linha como red (vermelho) e a espessura da linha como 2.
plt.figure(figsize=(8, 4))
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel('Valore real')
plt.ylabel('Valor predito')
plt.title('Dispersão dos dados')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.show()
```

```python
# Vamos ver quais atributos tiveram mais peso positivo ou negativo para o resultado do modelo.
# Vamos pegar os nomes das colunas de atributos e atribuir a uma variável#
nomes_atributos = x_train.columns
```

```python
nomes_atributos
```

```python
# Vamos criar um dataframe com o pandas, adicionamos os coeficientes do modelo, indicamos qual o nome dessa coluna que acabamos de criar e qual vai ser o nome das linhas, ou seja, dos index. E vamos armazenar o resultado em uma variável.
coefs = pd.DataFrame(model.coef_, columns=['coeficientes'], index= nomes_atributos)
```

```python
# Vamos ordenar essa coluna de coeficientes pra gente conseguir ver melhor o que ta influenciando mais ou menos. Fazemos isso usando a função de sort_values:
coefs = coefs.sort_values(by='coeficientes', ascending=False)
```

```python
# Vamos plotar um gráfico de barras horizontal para visualizar melhor e adicionamos uma linha na vertical no ponto 0 para termos como referência
coefs.plot.barh(figsize=(8, 6))
plt.axvline(x=0, color='red')
```
